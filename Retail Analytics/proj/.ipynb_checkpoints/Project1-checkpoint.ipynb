```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Define the function to clean and extract tags
def clean_and_extract_tags(text):
    doc = nlp(text.lower())
    tags = [token.text for token in doc if token.is_alpha and token.text not in STOP_WORDS]
    return ','.join(tags)

# Apply the function to each specified column
columns_to_extract_tags_from = ['Category', 'Brand', 'Description']

for column in columns_to_extract_tags_from:
    train[column] = train[column].apply(clean_and_extract_tags)
```

```python
train['Tags']=train[columns_to_extract_tags_from].apply(lambda row: ', '.join(row),axis=1)
```

# Function to Recommend Products for Content Base

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def content_based_recommendations(train_data, item_name, top_n=10):
    if item_name not in train_data['Name'].values:
        print(f"Item '{item_name}' not found in the training data.")
        return pd.DataFrame()

    vectorizer = TfidfVectorizer(stop_words='english')

    # Correcting the variable name here
    matrix_content = vectorizer.fit_transform(train_data['Tags'])

    similarities_content = cosine_similarity(matrix_content, matrix_content)

    item_index = train_data[train_data['Name'] == item_name].index[0]

    similar_items = list(enumerate(similarities_content[item_index]))

    similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)

    top_similar_items = similar_items[1:top_n+1]

    recommended_item_indices = [x[0] for x in top_similar_items]

    recommended_items_details = train_data.iloc[recommended_item_indices][['Name', 'ReviewCount', 'Brand', 'ImageURL', 'Rating']]

    return recommended_items_details
```

```python
# Example usage
item_name = 'Kokie Professional Matte Lipstick, Hot Berry, 0.14 fl oz'
content_based_rec = content_based_recommendations(train, item_name, top_n=8)
content_based_rec
```

# Collaborative Filtering

```python
user_item_matrix = train.pivot_table(index='ID', columns='ProdID', values='Rating',aggfunc='mean').fillna(0).astype(int)
```

```python
user_similarity = cosine_similarity(user_item_matrix)
```

```python
target_user_id = 4
target_user_index = user_item_matrix.index.get_loc(target_user_id)
```

```python
user_similarities = user_similarity[target_user_index]

similar_user_indices = user_similarities.argsort()[::-1][1:]
```

```python
recommend_items = []

for user_index in similar_user_indices:
    rated_by_similar_user = user_item_matrix.iloc[user_index]
    not_rated_by_target_user = (rated_by_similar_user==0) & (user_item_matrix.iloc[target_user_index]==0)
    
    recommend_items.extend(user_item_matrix.columns[not_rated_by_target_user][:10])

recommended_items_details = train[train['ProdID'].isin(recommend_items)][['Name','ReviewCount','Brand','ImageURL','Rating']]
```

# Function that Recommends Items

```python
def collaborative_filtering_recommendations(train_data, target_user_id, top_n=10):
    user_item_matrix = train.pivot_table(index='ID', columns='ProdID', values='Rating', aggfunc='mean').fillna(0)

    user_similarity = cosine_similarity(user_item_matrix)

    target_user_index = user_item_matrix.index.get_loc(target_user_id)

    user_similarities = user_similarity[target_user_index]

    similar_users_indices = user_similarities.argsort()[::-1][1:]

    recommended_items = []

    for user_index in similar_users_indices:
        rated_by_similar_user = user_item_matrix.iloc[user_index]
        not_rated_by_target_user = (rated_by_similar_user == 0) & (user_item_matrix.iloc[target_user_index] == 0)

        recommended_items.extend(user_item_matrix.columns[not_rated_by_target_user][:top_n])

    recommended_items_details = train[train['ProdID'].isin(recommended_items)][['Name', 'ReviewCount', 'Brand', 'ImageURL', 'Rating']]

    return recommended_items_details.head(10)

# Example usage
target_user_id = 4
top_n = 5
collaborative_filtering_rec = collaborative_filtering_recommendations(train, target_user_id)
print(f"Top {top_n} recommendations for User {target_user_id}:")
collaborative_filtering_rec
```

# Hybrid Recommendation

```python
def hybrid_recommendations(train_data,target_user_id, item_name, top_n=10):
    
    content_based_rec = content_based_recommendations(train_data,item_name, top_n)

    collaborative_filtering_rec = collaborative_filtering_recommendations(train_data,target_user_id, top_n)
    
    hybrid_rec = pd.concat([content_based_rec, collaborative_filtering_rec]).drop_duplicates()

    hybrid_rec = hybrid_rec.sort_values(by='Rating', ascending=False)
    
    return hybrid_rec.head(10)
```

```python
# Example usage
target_user_id = 4 
item_name = "OPI Nail Lacquer Polish .5oz/15mL - This Gown Needs A Crown NL U11" 
hybrid_rec = hybrid_recommendations(train,target_user_id, item_name, top_n=10)

print(f"Top 10 Hybrid Recommendations for User {target_user_id} and Item '{item_name}':")
hybrid_rec
```

```python
# Example usage: 
target_user_id = 10
item_name = 'Black Radiance Perfect Tone Matte Lip Crème, Succulent Plum'

hybrid_rec = hybrid_recommendations(train,target_user_id, item_name, top_n=10)

print(f"Top 10 Hybrid Recommendations for User {target_user_id} and Item '{item_name}':")
hybrid_rec
```

```python
# Example Usage
target_user_id = 4
item_name = "OPI Nail Lacquer Polish .5oz/15mL - This Gown Needs A Crown NL U11"

# Get recommendations from each method
content_based_rec = content_based_recommendations(train, item_name, top_n=10)[['ProdId', 'Name', 'Rating']]
collaborative_rec = collaborative_filtering_recommendations(train, target_user_id, top_n=10)[['ProdId', 'Name', 'Rating']]
hybrid_rec = hybrid_recommendations(train, target_user_id, item_name, top_n=10)[['ProdId', 'Name', 'Rating']]

# Display the results with only necessary columns
print(f"Content-Based Recommendations for Item '{item_name}':")
print(content_based_rec)

print(f"\nCollaborative Filtering Recommendations for User {target_user_id}:")
print(collaborative_rec)

print(f"\nHybrid Recommendations for User {target_user_id} and Item '{item_name}':")
print(hybrid_rec)
```
