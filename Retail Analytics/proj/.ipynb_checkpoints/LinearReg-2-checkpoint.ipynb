```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import and read dataset

```python
df = pd.read_csv("/content/Salary_Data.csv")
df
```

```python
df.shape
```

```python
x = df.loc[:, "YearsExperience"]
y = df.loc[:, "Salary"]
x
```

Split dataset

```python
X_train = x.iloc[:21,]
Y_train = y.iloc[:21,]
print(X_train)
print(Y_train)

X_test = x.iloc[21:,]
Y_test = y.iloc[21:,]
```

Calculate line equation

```python
N = len(X_train)
sumX = sum(X_train)
sumY = sum(Y_train)
sumXY = sum(X_train * Y_train)
sumXsq = sum(X_train ** 2)

b = ((N * sumXY) - (sumX * sumY)) / ((N * sumXsq) - (sumX ** 2))
a = (sumY - (b * sumX)) / N
```

```python
b
```

```python
a
```

Predict values

```python
def pred(a,b,x):
    return a + b*x
```

```python
for x in X_train:
  print("Experience : ",x,"Expected Salary:",pred(a,b,x))
```

```python
for x in X_test:
  print("Experience : ",x,"Expected Salary:",pred(a,b,x))
```

```python
y=pred(a,b,6)
y
```

```python
pred_test = pred(a,b,X_test)
pred_train = pred(a,b,X_train)
```

```python
plt.plot(X_train, pred_train, color="red")
plt.scatter(X_train, Y_train)
plt.show()
```

Plot graph

```python
plt.plot(X_test, pred_test, color="red")
plt.scatter(X_test, Y_test)
plt.show()
```

Calculate MSE

```python
def mean_squared_error(true, pred):
    squared_errors = (true - pred) ** 2
    mse = sum(squared_errors) / len(true)
    return mse
```

```python
mse_test = mean_squared_error(Y_test, pred_test)
mse_train = mean_squared_error(Y_train, pred_train)

print("MSE for Test Set:", mse_test)
print("MSE for Train Set:", mse_train)
```
